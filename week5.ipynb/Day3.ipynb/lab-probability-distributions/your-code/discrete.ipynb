````python
from scipy.stats import binom
n = 20
k = 5
p = 3/5

def calcular_probabilidad_binomial(n, p, k):
    probabilidad = binom.pmf(k, n, p)
    return probabilidad

result = calcular_probabilidad_binomial(n, p, k)
print(result)  
````

In the lesson we also discussed CDF (cumulative distribution function). In the context of the above problem, **what is the probability that less than 5 fruits picked are apples?**

Write the formula and calculate the probability below.

````python
n = 20  # Number of trials
p = 3/5  # Probability of success
# Create a binomial distribution object
binomial_dist = binom(n, p)

# Calculate the probability mass function (PMF) of specific outcomes
k = 5  # Number of successes
probability = binomial_dist.pmf(k)
print(f"The probability of getting {k} successes out of {n} trials is: {probability}")
# Calculate the cumulative distribution function (CDF)
k = 5  # Up to this number of successes
probability = binomial_dist.cdf(k)
print(f'The cdf of getting 5 or less is: {probability}')
````

````python
probability = binomial_dist.cdf(4)
print(f'The cdf of getting less than 5 is: {probability}')
````

**Plot the PDF (probability distributiom function) for the above binomial distribution.**

*Hint: The range of x in a binomial distribution is from `0` to `n`.*

````python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

n = 20
k = np.arange(n+1)  # Range of values from 0 to n
p = 3/5

# Calculate the probability for each value in k
probabilities = binom.pmf(k, n, p)

# Plot the PDF
plt.plot(k, probabilities, marker='o')
plt.xlabel('k')
plt.ylabel('PDF')
plt.title('Binomial Distribution (PDF)')
plt.grid(True)
plt.show()
````
