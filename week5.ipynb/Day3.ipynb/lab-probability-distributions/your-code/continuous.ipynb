## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
def normal(mu, sigma):
    return norm(mu, sigma)

a = normal(10,1)
b = normal(10,50)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_normal_numbers(mu, sigma, size):
    return np.random.normal(mu, sigma, size)

# Generating 1,000 normally distributed numbers with mean 10 and standard deviation 1
numbers_1 = generate_normal_numbers(10, 1, 1000)

# Generating 1,000 normally distributed numbers with mean 10 and standard deviation 50
numbers_2 = generate_normal_numbers(10, 50, 1000)

# Plotting the distributions
plt.figure(figsize=(10, 6))
plt.hist(numbers_1, bins=30, alpha=0.5, label='Mean=10, Std Dev=1')
plt.hist(numbers_2, bins=30, alpha=0.5, label='Mean=10, Std Dev=50')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Normal Distribution')
plt.legend()
plt.show()
```

How are the two distributions different?

```python
# They have different standard deviations
```

## Exponential Distribution

1. Plot the distributions as histograms with the number of bins as 100.

Your output should look like below:

```python
def exponential(mu, size):
    return np.random.exponential(scale= mu, size =size)

# mean is the same as lambda and we do the inverse
```

```python
a = exponential(1, 1000)
b = exponential(100, 1000)
```

```python
plt.hist(a, bins=100)
plt.show()

plt.hist(b, bins=100)
plt.show()
```

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e.  = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
from scipy.stats import expon

lambda_inv = 10
exp_distribution = expon(scale=lambda_inv)

p = exp_distribution.cdf(15)
```

What is the probability that the customer will spend more than 15 minutes

```python
1 - p
```
