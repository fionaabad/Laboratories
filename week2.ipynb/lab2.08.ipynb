## Exercise 1: Clean and Format the Data

```python
df1 = pd.read_csv('file1 (1).txt')
df2 = pd.read_csv('file2.txt')
df3 = pd.read_csv('file3.txt')
```

```python
def clean(df):
    df.columns = [col.replace(' ', '_').lower() for col in df.columns]
    
    def convert(cell):
        if str(cell) == 'nan':
            return 'nan'
        else:
            return float(cell.replace('%', ''))
    
    df['customer_lifetime_value'] = df['customer_lifetime_value'].apply(convert)
    df.gender = df.gender.replace('Male', 'M').replace('Femal', 'F').replace('female', 'F')
    df.customer_lifetime_value = df.customer_lifetime_value.astype(float)
    
    df = df.dropna(how='all')
    df.customer_lifetime_value.fillna(df.customer_lifetime_value.mean(), inplace=True)
    
    if df.duplicated().sum() == 0:
        pass
    
    return df
df1 = clean(df1)
df2= clean(df2)
```

```python
df3.columns = [col.replace(' ', '_').lower() for col in df3.columns]
df3.gender = df3.gender.replace('Male', 'M').replace('Femal', 'F').replace('female', 'F')
df3
```

```python
df1_sorted = df1.sort_values('customer')
df2_sorted = df2.sort_values('customer')
df3_sorted = df3.sort_values('customer')
```

```python
merged_id = pd.merge(df1, pd.merge(df2,df3, on= 'customer', how= 'outer'), on= 'customer', how ='outer')
merged_id
```
