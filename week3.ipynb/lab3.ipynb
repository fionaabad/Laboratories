### Plot both y= x**2 and y= e**x in the same plot using normal and logarithmic scale.

#### Normal

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the data
x = np.linspace(0, 5, 100)
y1 = x**2
y2 = np.exp(x)

# Create the figure and axes
fig, ax = plt.subplots()

# Plot the curves with normal scale
ax.plot(x, y1, label='y = x**2')
ax.plot(x, y2, label='y = e**x')

# Set labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Plots of y = x**2 and y = e**x (Normal Scale)')

# Add legend
ax.legend()

# Show the plot
plt.show()
```

#### Logarithmic

```python
# Create the figure and axes
fig2, ax2 = plt.subplots()

# Plot the curves with logarithmic scale on the y-axis
ax2.plot(x, y1, label='y = x**2')
ax2.plot(x, y2, label='y = e**x')

# Set labels and title
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('Plots of y = x**2 and y = e**x (Logarithmic Scale)')

# Set logarithmic scale on the y-axis
ax2.set_yscale('log')

# Add legend
ax2.legend()

# Show the plot
plt.show()
```

### Steps by month / active

```python
plt.rcParams['figure.figsize'] = [12,4]

monthly_median_steps = fitbit.groupby('Months_encoded')['Steps'].median()
df = monthly_median_steps.to_frame()

df.plot(kind='bar', color = 'blue')
plt.title('Steps per month', fontsize=15)

active = fitbit.groupby('Work_or_Weekend')['Steps'].median()
df1 = active.to_frame()
df1.plot(kind='bar', color = 'blue')
plt.title('Median steps Work or Weekend', fontsize=15)
```
